# 🪟 Interactive Tutorial: Sliding Window Pattern

Welcome to the interactive study guide for the **Sliding Window** pattern! 

This notebook will walk you through the concept using one of the most famous MAANG interview questions: **Longest Substring Without Repeating Characters**.

## 1. The Problem Statement

Given a string `s`, find the length of the **longest substring** without repeating characters.

**Example 1:**
- **Input:** `s = "abcabcbb"`
- **Output:** `3`
- **Explanation:** The answer is `"abc"`, with the length of 3.

## 2. The Naive Approach ($O(N^3)$)

The brute force way is to generate every possible substring, and for each one, check if it has any duplicate characters. 

This is terribly slow. Let's see how Sliding Window fixes this.

## 3. The Sliding Window Approach ($O(N)$)

Instead of restarting our substring search constantly, we can maintain a **window** of characters. 

1. We use two pointers: `left` and `right`.
2. We expand the window by moving `right` to the right.
3. If we encounter a character that is *already in our window*, we know the window is invalid. We shrink the window by moving `left` to the right until the duplicate is gone.

Let's write the code!

```python
def lengthOfLongestSubstring(s: str) -> int:
    char_map = {}
    left = 0
    max_len = 0
    
    print(f"Processing string: '{s}'\n")
    
    for right in range(len(s)):
        current_char = s[right]
        
        # If we've seen the character and it's inside our current window
        if current_char in char_map and char_map[current_char] >= left:
            print(f"⚠️ Found duplicate '{current_char}' at index {right}. Moving left pointer from {left} to {char_map[current_char] + 1}")
            left = char_map[current_char] + 1
            
        # Update the last seen index of the character
        char_map[current_char] = right
        
        # Calculate current window length and update max
        current_len = right - left + 1
        max_len = max(max_len, current_len)
        
        print(f"Step {right}: Window is '{s[left:right+1]}', Max Length so far: {max_len}")
        
    return max_len
```

## 4. Let's Test It Interactively!

Run the cell below to see how the window expands and shrinks in real-time.

```python
# Try changing this string to see how the algorithm reacts!
test_string = "abcabcbb"

result = lengthOfLongestSubstring(test_string)
print(f"\n🎉 Final Result: {result}")
```
